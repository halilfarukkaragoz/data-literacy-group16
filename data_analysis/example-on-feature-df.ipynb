```python
import pandas as pd 
import os
from pathlib import Path

import numpy as np
```

```python
pwd  = Path(os.getcwd())
pwd
```

```python
dataset_pkl = 'feature_df.pkl'

pkl_path = pwd.parent / "data" / dataset_pkl
df = pd.read_pickle(pkl_path)
df
```

```python
df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the DataFrame is named df
numerical_columns = [
    'title_length', 'abstract_length', 'tldr_length', 'author_count', 
    'keyword_count', 'metareview_length', 'review_count', 'review_confidence_nanmean', 
    'review_confidence_nanstd', 'review_confidence_nanmin', 'review_confidence_nanmax',
    'review_confidence_nanmedian', 'review_confidence_nansum', 'review_confidence_nanvar', 
    'review_confidence_nanprod', 'review_correctness_nanmean', 'review_correctness_nanstd', 
    'review_correctness_nanmin', 'review_correctness_nanmax', 'review_correctness_nanmedian', 
    'review_correctness_nansum', 'review_correctness_nanvar', 'review_correctness_nanprod',
    'review_technical_novelty_and_significance_nanmean', 
    'review_recommendation_nanmean', 'review_recommendation_nanstd',
    'review_recommendation_nanmin', 'review_recommendation_nanmax', 
    'review_recommendation_nanmedian', 'review_recommendation_nansum', 
    'review_recommendation_nanvar', 'review_recommendation_nanprod'
]

# Set up the matplotlib figure
plt.figure(figsize=(20, 40))

# Iterate over numerical columns to create a violin plot for each
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns) // 2, 2, i)  # Adjust number of rows/columns for subplots
    sns.violinplot(x='is_accepted', y=col, data=df, hue='is_accepted', legend=False)
    plt.title(f'Violin Plot of {col} by is_accepted')

plt.tight_layout()
plt.show()
```
