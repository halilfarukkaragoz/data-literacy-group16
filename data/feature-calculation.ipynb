```python
# calculating features for each paper

feature_df = paper_df.iloc[:, [0]].copy()
feature_df.rename(columns={'id': 'paper_id'}, inplace=True)
feature_df
```

```python
# maybe_important_fields = ['title', 'authors', 'keywords', 'TL;DR', 'abstract']

feature_df['title_length'] = paper_df['content'].apply(lambda x: len(x.get('title', '')))
feature_df['abstract_length'] = paper_df['content'].apply(lambda x: len(x.get('abstract', '')))
feature_df['tldr_length'] = paper_df['content'].apply(lambda x: len(x.get('TL;DR', '')))

feature_df['author_count'] = paper_df['content'].apply(lambda x: len(x.get('authors', [])))
feature_df['keyword_count'] = paper_df['content'].apply(lambda x: len(x.get('keywords', [])))

feature_df    
```

```python
# first we define the helper and main functions

import re
from tqdm import tqdm
tqdm.pandas()  


def _extract_numeric_prefix(maybe_string: object) -> int:
    """
    Extracts the leading integer from a string.
    If no integer is found, returns NaN.
    """
    string = str(maybe_string)
    match = re.match(r'^(\d+)', string.strip())
    return int(match.group(1)) if match else np.nan


def numeric_prefix_agg(series: 'pd.Series', fieldname: str, agg_func: callable):
    field_values = series.apply(lambda x: _extract_numeric_prefix(x.get(fieldname)) if isinstance(x, dict) else None)

    if field_values.isnull().all():
        return np.nan  # if cannot extract numeric prefix from any value, return NaN
    return agg_func(field_values)


# then we agg it
agg_functions_on_df = {
    'review_count': 'count',  # Count reviews
    # example entry. The dict will have (key,value) pairs like this:
    # 'recommendation_nanmean': lambda s: numeric_prefix_agg(s, 'rating', np.nanmean),
}


agg_functions = [np.nanmean, np.nanstd, np.nanmin, np.nanmax, np.nanmedian, np.nansum, np.nanvar, np.nanprod]
fields = ['confidence', 'correctness', 'technical_novelty_and_significance',
          'empirical_novelty_and_significance', 'recommendation']

# add the aggregation functions to the dictionary
for field in fields:
    for func in agg_functions:
        agg_functions_on_df[f'{field}_{func.__name__}'] = lambda s: numeric_prefix_agg(s, field, func)


# this is just 'review_df.groupby('replyto').agg(agg_functions_on_df).reset_index()' but I want tqdm. Thus the complexity.
review_agg = review_df.groupby('replyto').progress_apply(
    lambda group: group['content'].agg(agg_functions_on_df)
).reset_index()
review_agg
```

```python

review_agg.rename(columns={'replyto': 'paper_id'}, inplace=True)
feature_df = feature_df.merge(review_agg, on='paper_id', how='left')
feature_df
```
